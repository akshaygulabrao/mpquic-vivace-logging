```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("s2",names = ["s","ns","RTT","cwnd","lost"])
df.drop(df.index[df['RTT'] == 'GET 10.1.0.1:6121/random'], inplace = True)
df.drop(df.index[df['RTT'] == '0s'],inplace=True)
df["RTT"] = df["RTT"].str[:-2].astype('float64')
df.dropna(axis=0,inplace=True)
df["time"] = df["s"] + df["ns"]
```

```python
plt.style.use('default')
f = plt.figure()
ax1 = f.add_subplot(111)
ax1.plot(df["time"],df["cwnd"])
ax1.set_title('Latency(ms) and Loss(# pkts) vs cwnd(bytes)')
ax2 = f.add_subplot(111, sharex=ax1, frameon=False)
ax2.yaxis.tick_right()
ax2.plot(df["time"],df["RTT"],color = 'orange')
```

```python
f = plt.figure()
ax1 = f.add_subplot(111)
ax1.plot(df["time"],df["cwnd"])
ax2 = f.add_subplot(111, sharex=ax1, frameon=False)
ax2.yaxis.tick_right()
ax2.plot(df["time"],df["lost"],color = 'orange')
```
